# 1. Definição do Problema



## Descrição do Problema

Este projeto tem como objetivo analisar o impacto da Inteligência Artificial (IA) na vida estudantil na Índia, utilizando o dataset "AI_Student_Life_india.csv".

A crescente adoção de ferramentas de IA (como ChatGPT, assistentes inteligentes e plataformas adaptativas) tem transformado a forma como estudantes estudam, produzem trabalhos acadêmicos e organizam sua rotina.

O problema central é compreender:

- Como a IA está influenciando o desempenho acadêmico?
- Existe impacto no tempo de estudo?
- A IA melhora ou prejudica a aprendizagem?
- Há mudanças na saúde mental ou produtividade?

## Tipo de Problema

Trata-se inicialmente de um problema de **análise exploratória**, sem variável-alvo definida.

✅ PRÉ-PROCESSAMENTO



# 3. Pré-processamento de Dados

Objetivo:
- Limpar dados
- Preparar para modelagem futura
- Criar versões transformadas do dataset


🔹 Remoção de ID (não agrega informação)





```python
df = df.drop(columns=["Student_ID"])

```

🔹 One-Hot Encoding



```python
categorical_cols = df.select_dtypes(include="object").columns

encoder = OneHotEncoder(drop="first", sparse_output=False)

encoded = encoder.fit_transform(df[categorical_cols])

encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(categorical_cols)
)

df_encoded = pd.concat(
    [df.drop(columns=categorical_cols), encoded_df],
    axis=1
)

df_encoded.head()




```

🔹 Normalização





```python
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(
    scaler.fit_transform(df_encoded),
    columns=df_encoded.columns
)

df_normalized.head()
```

🔹 Padronização





```python
scaler_std = StandardScaler()

df_standardized = pd.DataFrame(
    scaler_std.fit_transform(df_encoded),
    columns=df_encoded.columns
)

df_standardized.head()

```
